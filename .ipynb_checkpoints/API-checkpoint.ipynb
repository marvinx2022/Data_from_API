El presente desafío consiste en generar la conexión con una API pública y crear un DataFrame para manipular con Python, luego realizar un EDA básico. 

Se elige la API de NREL, http://developer.nrel.gov/, de la cual se desea obtener el listado de empresas de distribución eléctrica del centro del estado de Texas en EEUU. 

Fue necesario obtener un acces token para el acceso a la información.

Se buscó una API con información que pueda ser de interés para el proyecto en proceso. 

```python
# Importamos librerías a utilizar

import numpy as np
import pandas as pd
import requests
import json 

from pandas import json_normalize

#Se toman Longitud: -99.2506100    Latitud: 31.2504400

#Se consideran 200 millas a la redonda desde el punto de coordenadas provisto, esto se obtiene de la documentación de la API. 

key= "9XssWsfhBWqSPoC71eiEKOX7hgQneW0OAQBWYDUW"

link='http://developer.nrel.gov/api/utility_rates/v3.json?api_key=9XssWsfhBWqSPoC71eiEKOX7hgQneW0OAQBWYDUW&lat=31.2504400&lon=-99.2506100&radius=200'

response=requests.get(link)
```

Se muestra el resultado obtenido:

```python
data=response.content
data
```

Se carga el resultado JSON:

```python
data = json.loads(data)
```

Se muestra la información:

```python
#Json obtenido
data
```

Se observa que los datos obtenidos "Output" se encuentran separados por el caracter "|" dentro del JSON. 

Se realiza un Split para luego ordenar los datos:

```python

result=data["outputs"]
result['company_id']
#separamos los valores para verificar en el campo "company_id"
company_id=result['company_id'].split("|")
utility_name=result['utility_name'].split("|")

type(utility_name)
```

```python
utility_name=np.array(utility_name)

company_id=np.array(company_id)


```

Se observa que se tiene un tipo JSON anidado, por lo que se buscará normalizar la información obtenida. 

```python
dfA=pd.DataFrame(utility_name)
dfA
```

```python
dfB=pd.DataFrame(company_id)
dfB
```

```python
result = pd.concat([dfB,dfA], axis=1)
result = result.set_axis(['Company_ID', 'Company_name'], axis=1)
result
```

### EDA básico de la información obtenida

Se realiza una exploración simple del Data Frame obtenido. 

Verificamos la cantidad de redes propiedad de cada empresa, para verificar las companias con mayor presencia en el sector:

```python
df1= result[['Company_name']].value_counts()
df1=pd.DataFrame(df1).set_axis(['Utilities_Qt'], axis=1)
df1.head(10)
```

Se obseva que existen dos companias propietarias de una gran cantidad de instalaciones compardas con el resto ya que concentran en total 49 registros de un total de 199 observados. Se presume que estas companias dominan el mercado, para apoyar esta hipotesis se podría agregar en una etapa futura la energía transportada o algun otro parametro que de indicios de la importancia estrategica real de las empresas en la zona. Se presenta la información en un histograma:

```python
import matplotlib as mpl
import seaborn as sns

frecuencias = result['Company_name'].value_counts()
top_10_categorias = frecuencias.head(10).index
df_top_10 = result[result['Company_name'].isin(top_10_categorias)]

plt.figure() 
histograma = sns.histplot(df_top_10['Company_name'], bins=len(top_10_categorias), color='skyblue')

plt.xlabel('Propietarios')
plt.ylabel('Cantidad de redes')
plt.title('Histograma de los 10 propietarios de mayor importancia')

plt.xticks(rotation=45, ha='right')
plt.show()
```
